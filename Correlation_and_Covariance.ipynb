Define Covariance and explain how it differs from Correlation in terms of scale and interpretation.
->  Covariance measures how two variables change together: positive if both rise/fall together, negative if opposite, zero if no pattern

Covariance:  Can be any number (-∞ to +∞)
Has units (like kg×cm)
Shows direction only

Correlation: Always between -1 and +1
 No units
 Shows direction AND strength

What does a positive, negative, and zero covariance indicate about the relationship between two variables?
Negative Covariance:
Two things move in the same direction together.
When one goes up, the other goes up too. When one goes down, the other goes down.

Zero Covariance:
No connection between them.
One changing doesn't affect the other at all.



 Discuss the limitations of covariance as a measure of relationship between two variables. Why is correlation
preferred in many cases?

Covariance changes if you change units. Height in cm vs inches = different covariance values. Can't compare across datasets.
One extreme value can completely change the covariance.

Pearson's Correlation:
Measures straight-line linear relationship
Uses actual data values
Assumes normal distribution
Spearman's Correlation:

Measures rank order monotonic relationship
Converts data to ranks first
No normality assumption

If the correlation coefficient between two variables X and Y is 0.85, interpret this value in context. Can you

A correlation of 0.85 indicates a strong positive linear relationship between X and Y.
When X increases, Y increases consistently
85% of variation in one explains the other
